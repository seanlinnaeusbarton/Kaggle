```python
import numpy as np
import pandas as pd
pd.set_option("precision",16)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%%time
data = pd.read_csv("./train/train.csv", dtype={'acoustic_data':np.int16, 'time_to_failure':np.float32})
```

```python
t = np.arange(0,len(data))
short_t = t[::50]
short_acoustic = data['acoustic_data'].values[::50]
short_failure = data['time_to_failure'].values[::50]

T = np.abs(short_failure[0]-short_failure[1])
#short_t = np.arange(0,len(short_acoustic),T)
#print(T)

#t = np.arange(0,len(short_acoustic))
fig, ax1 = plt.subplots()
ax1.plot(short_t,short_acoustic)
ax1.set_xlabel('time')
ax1.set_ylabel('acoustic signal')

ax2 = ax1.twinx()
ax2.plot(short_t,short_failure, 'r')

del t
del short_t
del short_acoustic
del short_failure
```

```python
zero_index = np.argmin(data['time_to_failure'].values[:])
acoustic = data['acoustic_data'].values[41:int(400)]
acoustic = acoustic - np.mean(acoustic)
d_acoustic = np.diff(acoustic)
failure = data['time_to_failure'].values[41:int(399)]
t = np.arange(0,len(acoustic))

fig, ax1 = plt.subplots()
ax1.plot(t[1:],acoustic[1:])
ax1.plot(t[1:],d_acoustic, 'g')
ax1.set_xlabel('time')
ax1.set_ylabel('acoustic signal')

ax2 = ax1.twinx()
ax2.plot(t[1:],failure, 'r')

non_zero = np.nonzero(np.diff(failure))

N = len(acoustic)
T = np.abs(failure[non_zero[0][0]]-failure[non_zero[0][0]+1])

np.diff(non_zero)
```

```python
f = np.linspace(0, 1/T, N-1)

ft = np.fft.fft(d_acoustic)
ft_freq = np.fft.fftfreq(N-1, d=T)

fig, ax = plt.subplots()
ax.plot(ft_freq[0:N // 2], np.abs(ft)[0:N // 2] * 1 / N)
ax.ylabel="Amplitude"
ax.xlabel="Frequency (Hz)"

```
